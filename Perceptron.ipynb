```python
#Import library
import numpy as np
```

```python
#Defining activation
def activation_step(x, threshold =0):
  return 1 if x >threshold else 0
```

```python
#Defining Perceptron model

class Perceptron:
    def __init__(self, input_size):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return activation_step(summation)

    def train(self, inputs, target_output, learning_rate=0.1, epochs=100):
        for epoch in range(epochs):
            total_error = 0
            for input_data, target in zip(inputs, target_output):
                prediction = self.predict(input_data)
                error = target - prediction
                total_error += abs(error)
                # print(f"error is {total_error}")
                self.weights += learning_rate * error * input_data
                self.bias += learning_rate * error
            if total_error == 0:
                break
```

NOT Gate

```python
inputs = np.array([[0],[1]])
target = np.array([1,0])
```

```python
print(inputs.shape)
```

```python
not_gate = Perceptron(inputs.shape[-1])
```

```python
not_gate.train(inputs, target)
```

```python
print("Testing Perceptron NOT gate:")
for input_data in inputs:
    output = not_gate.predict(input_data)
    print(f"Input: {input_data}, Output: {output}")
```

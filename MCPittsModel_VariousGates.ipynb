Initialise library

```python
import numpy as np
```

I is the input vector and W is the weight vector

```python
I = np.random.choice([0,1],3)
W = np.random.choice([-1,1],3)
```

printing the vectors

```python
print('Input vector',I)
print('Weight vector is',W)
```

dot product of input and weight vector

```python
# dot_prod = I * W
dot_prod = I @ W
print('dot product is', dot_prod)
```

defining the threshold gate

```python
def threshold_gate(dot_prod, T):
  if dot_prod >=T:
    print('Activated')
    return 1
  else:
    print('Not activated')
    return 0
```

Activating the gate based on the dot product

```python
T = 1
activation = threshold_gate(dot_prod,T)
print('Activation',activation)
```

##AND Gate

```python
input_table = np.array([
    [0,0],
    [0,2],
    [2,0],
    [2,2]
])
```

```python
#weights = np.random.choice([-1,1],2)
weights = np.array([2,2])
```

```python
dot_prods = input_table @ weights
print('Dot product is', dot_prods)
```

```python
T = 8
for ii in range(0,4):
  activation = threshold_gate(dot_prods[ii],T)
  print('Activation', activation)
```

##OR Gate

```python
T = 1
for ii in range(0,4):
  activation = threshold_gate(dot_prods[ii],T)
  print('Activation:', activation)
```

##XOR Gate

```python
def xor_gate(input_1, input_2):
  weights = np.array([1,1])
  T = 2
  comb_init = np.array([input_1, input_2])
  dot_p = comb_init @ weights
  activation = threshold_gate(dot_p,T)
  print(f'XOR ({input_1},{input_2}) = {activation}')
  return activation
```

```python
for input_1 in [0,1]:
  for input_2 in [0,1]:
    result = xor_gate(input_1, input_2)
```
